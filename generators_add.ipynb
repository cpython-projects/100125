**Задача 1.** Анализ логов  
Файл logs.txt содержит строки в формате:  
2024-11-20 12:45:23 INFO User logged in  
2024-11-20 12:47:01 ERROR Failed to load resource  
...  
Напишите генератор, который будет читать файл построчно и возвращать только строки с уровнем ERROR.  

**Задача 2.** Агрегация данных о продажах  
В текстовом файле sales.txt каждая строка представляет собой запись о продаже:  
{"region": "North", "sales": 2500}  
{"region": "South", "sales": 3000}  
Создайте генератор, который будет суммировать продажи для каждой строки и возвращать только общую сумму.  

**Задача 3.** Нормализация JSON-данных  
В JSON-файле data.json хранится массив объектов:  
```
[
    {"name": "Alice", "score": 95},
    {"name": "Bob", "score": 88},
    {"name": "Charlie", "score": null}
]
```
Напишите генератор, который будет перебирать объекты и заменять значения null на 0, возвращая преобразованные записи.

**Задача 4.** Поиск уникальных слов  
Дан текстовый файл text.txt. Создайте генератор, который будет читать строки из файла и возвращать уникальные слова из каждой строки (без учета регистра).  

**Задача 5.** Фильтрация данных  
В файле data.json хранится массив объектов:  
```
[
    {"id": 1, "value": 50},
    {"id": 2, "value": 70},
    {"id": 3, "value": 30}
]
```
Создайте генератор, который будет перебирать элементы JSON, возвращая только те объекты, где value > 40.

**Задача 6.** Объединение данных  
Даны два файла file1.txt и file2.txt. Напишите генератор, который построчно объединяет строки из двух файлов, возвращая пары строк в формате (строка из file1, строка из file2).  

**Задача 7.** Генерация последовательности событий  
На вход генератору передается список временных меток:  
timestamps = ["2024-11-20 12:00", "2024-11-20 13:00", "2024-11-20 14:30"]  
Напишите генератор, который создает пары последовательных событий в формате:  
("2024-11-20 12:00", "2024-11-20 13:00")  
("2024-11-20 13:00", "2024-11-20 14:30")  

**Задача 8.** Поиск аномалий в данных  
У вас есть список температур за месяц:  
temperatures = [22, 24, 21, 45, 23, 22, 20, 50, 22]  
Напишите генератор, который возвращает только те значения, которые выходят за пределы интервала [20, 30].  

**Задача 9.** Сегментация клиентов  
Дан список клиентов:  
```
customers = [
    {"id": 1, "revenue": 1000},
    {"id": 2, "revenue": 500},
    {"id": 3, "revenue": 7000},
]
```
Создайте генератор, который возвращает клиентов, разделенных на сегменты: Low, Medium, High в зависимости от их дохода (revenue).

**Задача 10.** Распределение данных по интервалам  
Дан список возрастов пользователей:  
ages = [23, 45, 35, 19, 60, 72, 50, 34]  
Напишите генератор, который возвращает распределение пользователей по возрастным интервалам (например, "18-25", "26-35").  

**Задача 11.** Умная фильтрация дат  
Вам дали список дат:  
dates = ["2024-01-01", "2024-11-20", "2024-12-25", "2025-01-01"]  
Напишите генератор, который возвращает только те даты, которые находятся в пределах следующего месяца.
