```python
# python hw_26.py /home/user/documents

import sys
import os

_, dir_path = sys.argv # ['hw_26.py', '/home/user/documents']

dirs = []
files = []

for obj_name in os.listdir(dir_path):
    if os.path.isdir(obj_name):
        dirs.append(obj_name)
    else:
        files.append(obj_name)

print('Folders:')
print(*dirs, sep='\n')

print('Files:')
print(*files, sep='\n')
```

```python
dir_path = os.getcwd()

dirs = []
files = []

for obj_name in os.listdir(dir_path):
    if os.path.isdir(obj_name):
        dirs.append(obj_name)
    else:
        files.append(obj_name)

print('Folders:')
print(*dirs, sep='\n')

print('\n', '-' * 20, '\n')

print('Files:')
print(*files, sep='\n')
```

```python
from collections import defaultdict

dir_path = os.getcwd()

dirs_files = defaultdict(list)

for obj_name in os.listdir(dir_path):
    dirs_files[os.path.isdir(obj_name)].append(obj_name)
    
print('Folders:')
print(*dirs_files[True], sep='\n')

print('\n', '-' * 20, '\n')

print('Files:')
print(*dirs_files[False], sep='\n')
```
