### Условие:

Создайте класс `User`, представляющий пользователя некоторого сайта.

---

### Атрибуты экземпляра:

* `first_name`: имя пользователя (первая часть поля `name`)
* `last_name`: фамилия пользователя (последняя часть поля `name`)
* `email`: из поля `email`
* `country`: используйте `address.city` как суррогат страны
* `birthdate`: случайно сгенерированная дата рождения в формате `YYYY-MM-DD`, сохраняется как `datetime.date`

---

### Поведение класса:

1. Строковое представление (`__str__`) выводит:

   ```
   Имя Фамилия, email, страна, дата рождения
   ```

2. Классовое поле `count`: количество всех созданных пользователей.

3. Классовый метод `total()` — возвращает общее количество пользователей.

4. Статический метод `permission(users, age)`:

   * Принимает список объектов `User` и целое число `age`.
   * Возвращает список пользователей, достигших указанного возраста на текущую дату.

---

### Используйте API:

* Получите 10 пользователей с `https://jsonplaceholder.typicode.com/users`.
* Сгенерируйте для каждого случайную дату рождения в диапазоне:

  * от 13 до 45 лет назад (от сегодняшней даты).
* Создайте список из объектов `User`.

---

### Выведите:

1. Данных пользователей, у которых есть доступ к ресурсу с возрастным ограничением **18+**.
2. Данных пользователей, у которых есть доступ к ресурсу с возрастным ограничением **25+**.
3. Общее количество пользователей.

---

### Дополнительно:

Используя данные из `/posts` или `/todos`:

* Добавить метод `load_posts(user_id)` — загружает список постов пользователя.
* Подсчитать активность пользователя по количеству постов или выполненных задач.
